In my RBC dataset analysis, the Simple Linear Regression model failed due to its limited predictive power on non-linear data. The Polynomial Linear Regression model performed better by capturing the dataset's non-linear relationships, enhancing prediction accuracy.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv("RBC.csv")
X = data_set.iloc[ : , :-1].values
y = data_set.iloc[:, -1].values
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X, y)
```

```python
plt.scatter(X, y)
plt.plot(X, regressor.predict(X), color = "red")
plt.title("RBC - Linear Regression", fontsize = 20)
plt.xlabel("Age", fontsize = 15)
plt.ylabel("RBC", fontsize = 15)
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
regressor2 = LinearRegression()
regressor2.fit(X_poly, y)
```

```python
plt.scatter(X, y)
plt.plot(X, regressor2.predict(X_poly), color = "black")
plt.title("RBC - Polynomial Regression", fontsize = 20)
plt.xlabel("Age", fontsize = 15)
plt.ylabel("RBC", fontsize = 15)
plt.show()
```

```python
# Predicting new result with simple linear regression
# Huge variation in predicting the output

regressor.predict([[7]])
```

```python
# Predicting new result with simple linear regression
# Accurate prediction

regressor2.predict(poly_reg.fit_transform([[7]]))
```
